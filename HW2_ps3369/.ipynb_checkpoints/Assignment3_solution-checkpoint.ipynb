```python
from __future__ import division, print_function
__author__ = 'ps3369 CUSP 2016'
import os
import numpy as np
from numpy.random import randn
import pylab as pl

from pandas import Series, DataFrame
import pandas
import pandas as pd

```

```python
DFdata = os.getenv("DFDATA")
print(DFdata)
```

```python
df_total = pd.read_csv(DFdata + "/ic3t-wcy2/1473437392/ic3t-wcy2.csv").head(20)
df_total
```

```python
df_total  = df_total[["Block","Job #"]]
df_total.head(20)
```

```python
df = pd.DataFrame(df_total[["Block","Job #"]], columns=['Block','Job #'])

df.plot.scatter(x='Block', y='Job #', color='red', label='Job # vs Block');

plt.title('Final Scatter Plot', fontsize = 20 , color='Blue')

plt.text(-2000, -150000, 'Fig: Block vs Job Number plotted', fontsize = 15)

plt.show()
```
